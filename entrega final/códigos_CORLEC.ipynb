```python
## lectura del archivo y división del texto a partir de los saltos de línea
file=open ('corpus_CORLEC.txt', 'r', encoding="utf-8")  
text=file.read()
texto=text.split("\n")
```

```python
print(texto[:30])
```

```python
import re
```

```python
##creo una nueva lista donde voy a guardar el texto nuevo limpio de algunas expresiones que constituyen encabezados o metatexto
texto_limpio =[]
for oracion in texto:
    expresion_por_reemplazar1 = re.compile(r'#.*$') ##esto va a buscar las líneas que contienen metadatos de la grabación
    texto_limpio += [expresion_por_reemplazar1.sub("", oracion)]
print(texto_limpio[:20])
```

```python
texto_limpio2 =[]
for oracion in texto_limpio:
    expresion_por_reemplazar2 = re.compile(r"|TEXTSEPARATOR|") ##esto va a buscar la expresión "textseparator"
    texto_limpio2 += [expresion_por_reemplazar2.sub("", oracion)]
print(texto_limpio2[:20])
```

```python
texto_limpio3 =[]
for oracion in texto_limpio2:
    expresion_por_reemplazar3 = re.compile(r"H[0-9a-zA-Z]+:\s") ##esto va a buscar las indicaciones de quién habla antes de su enunciado
    texto_limpio3 += [expresion_por_reemplazar3.sub("", oracion)]
print(texto_limpio3[:120])
```

```python
texto_limpio4 =[]
for oracion in texto_limpio3:
    expresion_por_reemplazar4 = re.compile(r"\[[a-zA-Z0-9_ ñáéíóúÑÁÉÍÓÚ]*\]") ##esto va a buscar las anotaciones del transcriptor, entre corchetes
    texto_limpio4 += [expresion_por_reemplazar4.sub("", oracion)]
print(texto_limpio4[:100])
```

```python
texto_limpio5 =[]
for oracion in texto_limpio4:
    expresion_por_reemplazar5 = re.compile(r"\.\.\.") ##esto elimina los tres puntos
    texto_limpio5 += [expresion_por_reemplazar5.sub("", oracion)]
print(texto_limpio5[:100])
```

```python
##armo una lista y guardo alli las oraciones en que divido el texto. Necesito usar un for para ingresar a cada salto de línea de la lista "textolimpio5"
oraciones=[]
for oracion in texto_limpio5:
    oraciones += oracion.split(".") 
print(oraciones[:25])
##ahora tengo una lista de oraciones
```

```python
##elimino espacios sobrantes
oraciones_limpio=[]
for oracion in oraciones:
    oraciones_limpio += [oracion.strip()]
print(oraciones_limpio[:100])
```

```python
## elimino strings vacios
oraciones_limpio1=[]
for oracion in oraciones_limpio:
    if len(oracion)>0:
        oraciones_limpio1 += [oracion]
print(oraciones_limpio1[:100])
```

```python
import string

## ponemos todo en minúsculas
oraciones_min=[]
for palabra in oraciones_limpio1:
    oraciones_min+= [palabra.lower()]
print(oraciones_min[:20])
```

```python
##eliminamos primera línea
oraciones_min1 = oraciones_min[1:-1]
oraciones_min1[:20]
```

```python
## creo una función para eliminar puntuación
def remover_puntuacion(s): 
    for c in string.punctuation:
        s=s.replace(c,"")
    return s
```

```python
## armo lista "oraciones_min_sin_punt" para guardar allí las oraciones sin puntuación
oraciones_min_sin_punt =[]
for oracion in oraciones_min1:
    oraciones_min_sin_punt += [remover_puntuacion(oracion)]
```

```python
## reviso si funciona el código con un print
print(oraciones_min_sin_punt[:20])
```

```python
##remuevo signo de pregunta y de interrogación inicial, que no se elimina con string.puntuacion
oraciones_min_sin_punt1 =[]
oraciones_min_sin_punt2 =[]
for oracion in oraciones_min_sin_punt:
    oraciones_min_sin_punt1 += [oracion.replace("¿","")]
for oracion in oraciones_min_sin_punt1:
    oraciones_min_sin_punt2 += [oracion.replace("¡","")]
oraciones_min_sin_punt2[:20]
```

```python
##elimino espacios sobrantes
## creo una lista que contenga las oraciones limpias de doble espacio "oraciones_final"
oraciones_final =[]
for oracion in oraciones_min_sin_punt2:
    espacios_dobles = re.compile(r'\s\s') ##busca dos espacios en blanco seguidos
    oraciones_final += [espacios_dobles.sub(" ", oracion)]
print(oraciones_final[:20])
```

```python
## cambiamos el nombre de la primera columna a "posicion"
verbos_y_negacion_posicion.rename(columns={'Unnamed: 0': 'Posicion'}, inplace=True)
verbos_y_negacion_posicion
```

```python
##armamos una lista con todas las posiciones para poder extraer luego las sucesivas
lista_posiciones = list(verbos_y_negacion_posicion['Posicion'])
```

```python
##creamos una función para determinar posiciones sucesivas (dos o más verbos/aux/adv negados que se encuentren sucesivos))
def ranges(nums):
    ## ordena los números de menor a mayor sin repeticiones
    nums = sorted(set(nums))
    ## con esto se encuentran los gaps
    gaps = [[s, e] for s, e in zip(nums, nums[1:]) if s+1 < e]
    edges = iter(nums[:1] + sum(gaps, []) + nums[-1:])
    return [pair for pair in list(zip(edges, edges)) if pair[0]!=pair[1]]
```

```python
ranges(lista_posiciones)
```

```python
## hacemos una función para extraer toda la información de estas posiciones sucesivas
def construcciones_negacion_dict(rango):
    construcciones_negacion = verbos_y_negacion_posicion[verbos_y_negacion_posicion['Posicion'].isin(list(range(rango[0],rango[1]+1)))]
    return construcciones_negacion.to_dict('list')
```

```python
## armamos una lista que contenga toda la información de estas posiciones sucesivas
lista_construcciones_negacion = []

for rango in ranges(lista_posiciones):
    lista_construcciones_negacion += [construcciones_negacion_dict(rango)]
```

```python
lista_construcciones_negacion[:100]
```

```python
## hacemos una lista con lo que necesitamos eliminar de la lista de construcciones, ya que no constituyen construcciones multiverbales
paraeliminar=[]
for dicc in lista_construcciones_negacion: 
    if dicc["POS"]==["ADV","AUX"]: ## negaciones en primera posición más un único elemento verbal
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["POS"]==["ADV","VERB"]: ## negaciones en primera posición más un único elemento verbal
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["POS"]==["AUX","ADV"]: ## un único elemento verbal más negación en última posición 
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["POS"]==["VERB","ADV"]: ## un único elemento verbal más negación en última posición 
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["POS"]==["ADV","ADV"]: ## solo adverbios
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["POS"]==["ADV","ADV", "ADV"]: ## solo adverbios
        paraeliminar+=[dicc]
for dicc in lista_construcciones_negacion:
    if dicc["Lema"][0]=="ser" and dicc["Morfologia"][1]=="VerbForm=Inf": ## verbo ser + infinitivo (p.e. "es decir", sea llamar", "era fichar")
        paraeliminar+=[dicc]
paraeliminar
```

```python
## eliminamos las que podrían haber quedado como construcciones de una sola palabra
paraeliminar2=[]
for dicc in lista_construcciones_negacion: 
    if len(dicc["POS"])==0|1: 
         paraeliminar2+=[dicc]
paraeliminar2
```

```python
## creamos una función para eliminar una lista de dentro de otra
def eliminarlista(a,b):
    for i in range(len(b)):
        while b[i] in a:
            a.remove(b[i])
```

```python
## eliminamos todos los diccionarios que definimos para eliminar de nuestra lista de construcciones
eliminarlista(lista_construcciones_negacion,paraeliminar)
eliminarlista(lista_construcciones_negacion,paraeliminar2)
```

```python
##eliminamos las negaciones iniciales y finales de las construcciones multiverbales, para conservar solo las interpuestas
for dicc in lista_construcciones_negacion:
    if dicc["POS"][0]=="ADV":
        del dicc["Posicion"][0]
        del dicc["Palabra"][0]
        del dicc["Lema"][0]
        del dicc["POS"][0]
        del dicc["Morfologia"][0]
for dicc in lista_construcciones_negacion:
    if dicc["POS"][-1]=="ADV":
        del dicc["Posicion"][-1]
        del dicc["Palabra"][-1]
        del dicc["Lema"][-1]
        del dicc["POS"][-1]
        del dicc["Morfologia"][-1]
lista_construcciones_negacion[:20]
```

```python
## calculamos cantidad de auxiliares
aux = []
valAux = "AUX"
for dicc in construcciones_filtradas:
    for k, v in dicc.items():
        for POS in v:
            if valAux==POS:
                aux.append(valAux)
print(len(aux))
```

```python
## calculamos cantidad de construcciones de gerundio, infinitivo y participio (pueden superponerse en una misma secuencia)
Inf = []
valInf = "VerbForm=Inf"
for dicc in construcciones_filtradas:
    for k, v in dicc.items():
        for morf in v:
            if valInf==morf:
                Inf.append(valInf)
print(len(Inf))
```
